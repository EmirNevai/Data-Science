```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

df = pd.read_csv("df.csv")
df.head()
```

```python
print(df.Country.unique())
print(df.Year.unique())
```

```python
life_byyear = df.groupby("Year")["Life expectancy at birth (years)"].mean().reset_index()

plt.subplot(2, 1, 1)
plt.plot(life_byyear["Year"], life_byyear["Life expectancy at birth (years)"])
plt.show()
plt.clf()
```

```python
df["Country"] = df["Country"].replace("United States of America", "USA")
sns.barplot(x="Country", y="Life expectancy at birth (years)", data=df)
plt.show()
plt.clf()
```

```python
countries = df["Country"]
country_labels = df.Country.unique()
for country in countries:
    sns.scatterplot(x="Year", y="Life expectancy at birth (years)", data=df[df["Country"] == country])
plt.legend(country_labels, loc=4, fontsize="x-small")
```

```python
sns.lineplot(x="Year", y="Life expectancy at birth (years)", data=df, hue="Country")
plt.legend(bbox_to_anchor=(1, 0.5))
```

```python
mexico_data = df[df["Country"] == "Mexico"]
chile_data = df[df["Country"] == "Chile"]
china_data = df[df["Country"] == "China"]
zimbabwe_data = df[df["Country"] == "Zimbabwe"]
germany_data = df[df["Country"] == "Germany"]
usa_data = df[df["Country"] == "USA"]

plt.subplot(3, 2, 1)
plt.plot(chile_data["Year"], chile_data["Life expectancy at birth (years)"])

plt.subplot(3, 2, 2)
plt.plot(germany_data["Year"], germany_data["Life expectancy at birth (years)"])

plt.subplot(3, 2, 3)
plt.plot(mexico_data["Year"], mexico_data["Life expectancy at birth (years)"])

plt.subplot(3, 2, 4)
plt.plot(usa_data["Year"], usa_data["Life expectancy at birth (years)"])

plt.subplot(3, 2, 5)
plt.plot(zimbabwe_data["Year"], zimbabwe_data["Life expectancy at birth (years)"])

plt.show()
```

```python
print("Chile: " + str(round(chile_data["Life expectancy at birth (years)"].mean(), 2)))
print("China: " + str(round(china_data["Life expectancy at birth (years)"].mean(), 2)))
print("Germany: " + str(round(germany_data["Life expectancy at birth (years)"].mean(), 2)))
print("Mexico: " + str(round(mexico_data["Life expectancy at birth (years)"].mean(), 2)))
print("Usa: " + str(round(usa_data["Life expectancy at birth (years)"].mean(), 2)))
print("Zimbabwe: " + str(round(zimbabwe_data["Life expectancy at birth (years)"].mean(), 2)))
```
